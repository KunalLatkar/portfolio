# Removing null values

```python
most_common_location = df["Monitoring Location"].mode().iloc[0]

df["Monitoring Location"].fillna(most_common_location, inplace = True)

print(df["Monitoring Location"].isnull().sum())
```

```python
most_common_location = df["Station Code"].mode().iloc[0]

df["Station Code"].fillna(most_common_location, inplace = True)

print(df["Station Code"].isnull().sum())
```

```python
import numpy as np

df["State"] = df['State'].astype(str).replace(" ", np.nan)

prev_index = df["State"].shift(1)
next_index = df["State"].shift(-1)

value = df["State"].isna() & (prev_index == next_index)

df.loc[value, "State"] = prev_index[value]

print(df["State"].isnull().sum())
```

```python
df["Temperature (°C) Max"] = pd.to_numeric(df["Temperature (°C) Max"], errors='coerce')

df["State"] = df['State'].astype(str)

df['Temperature (°C) Max'] = df.groupby('State')["Temperature (°C) Max"].transform( lambda x : x.fillna(x.mean()))

print(df['Temperature (°C) Max'].isnull().sum())
```

Using IQR (Interquartile Range)
Outliers are data points outside the range:

*   𝑄1 − 1.5 × 𝐼𝑄𝑅,
*   𝑄3 + 1.5 × 𝐼𝑄𝑅
*   IQR = Q3 - Q1





```python
# error : TypeError: Cannot perform 'ror_' with a dtyped [float64] array and scalar of type [bool]
#Following to lines are to check for this error

#print(df['Total Coliform (MPN/100ml) Max'].dtype)
#print(df['Total Coliform (MPN/100ml) Max'].apply(type).value_counts())


Q1 = df['Total Coliform (MPN/100ml) Max'].quantile(0.25)
Q3 = df['Total Coliform (MPN/100ml) Max'].quantile(0.75)

IQR = Q3 - Q1

lower_bound =  Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

outliers_coliform = df[(df['Total Coliform (MPN/100ml) Max'] < lower_bound) |
                       (df['Total Coliform (MPN/100ml) Max'] > upper_bound)]

print(f"outliers of coliform: {outliers_coliform['Total Coliform (MPN/100ml) Max'].count()}")
```

Applying IQR on each numeric field

```python
num_fields = ["Temperature (°C) Min", "Temperature (°C) Max", "Dissolved Oxygen (mg/L) Min", "Dissolved Oxygen (mg/L) Max", "pH Min", "pH Max",
              "Conductivity (µmho/cm) Min", "Conductivity (µmho/cm) Max", "BOD (mg/L) Min", "BOD (mg/L) Max", "Nitrate N (mg/L) Min", "Nitrate N (mg/L) Max",
              "Fecal Coliform (MPN/100ml) Min", "Fecal Coliform (MPN/100ml) Max","Total Coliform (MPN/100ml) Min", "Total Coliform (MPN/100ml) Max",
              "Fecal Streptococci (MPN/100ml) Min", "Fecal Streptococci (MPN/100ml) Max", "Year" ]

for field in num_fields:
  Q1 = df[field].quantile(0.25)
  Q3 = df[field].quantile(0.75)

  IQR = Q3 - Q1

  lower_bound =  Q1 - (1.5 * IQR)
  upper_bound = Q3 + (1.5 * IQR)

  outliers_coliform = df[(df[field] < lower_bound) |
                        (df[field] > upper_bound)]

  print(f"outliers of {field}: {outliers_coliform[field].count()}")
```

using median to correcting outliers

```python

num_fields = ["Temperature (°C) Min", "Temperature (°C) Max", "Dissolved Oxygen (mg/L) Min", "Dissolved Oxygen (mg/L) Max", "pH Min", "pH Max",
              "Conductivity (µmho/cm) Min", "Conductivity (µmho/cm) Max", "BOD (mg/L) Min", "BOD (mg/L) Max", "Nitrate N (mg/L) Min", "Nitrate N (mg/L) Max",
              "Fecal Coliform (MPN/100ml) Min", "Fecal Coliform (MPN/100ml) Max","Total Coliform (MPN/100ml) Min", "Total Coliform (MPN/100ml) Max",
              "Fecal Streptococci (MPN/100ml) Min", "Fecal Streptococci (MPN/100ml) Max", "Year" ]

for field in num_fields:
    Q1 = df[field].quantile(0.25)
    Q3 = df[field].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    median_value = df[field].median()

    # Only replace values that are outliers
    outlier_mask = (df[field] < lower_bound) | (df[field] > upper_bound)
    df.loc[outlier_mask, field] = median_value
```

# Adding Target variable

creating a field that can act a target varible(contamination)

```python
num_fields = ["Temperature (°C) Min", "Temperature (°C) Max", "Dissolved Oxygen (mg/L) Min", "Dissolved Oxygen (mg/L) Max", "pH Min", "pH Max",
              "Conductivity (µmho/cm) Min", "Conductivity (µmho/cm) Max", "BOD (mg/L) Min", "BOD (mg/L) Max", "Nitrate N (mg/L) Min", "Nitrate N (mg/L) Max",
              "Fecal Coliform (MPN/100ml) Min", "Fecal Coliform (MPN/100ml) Max","Total Coliform (MPN/100ml) Min", "Total Coliform (MPN/100ml) Max",
              "Fecal Streptococci (MPN/100ml) Min", "Fecal Streptococci (MPN/100ml) Max", "Year" ]

for field in num_fields:
  print(f"range of {field}: {df[field].min()}, {df[field].max()}")
```

| Parameter                          | Safe Range / Limit                                                                                     |
| ---------------------------------- | ------------------------------------------------------------------------------------------------------ |
| **Temperature (°C)**               | No strict limit, but >35°C may be problematic (skip or use informally)                                 |
| **Dissolved Oxygen (mg/L)**        | **≥ 5.0 mg/L** (Min value must be ≥ 5.0)                                                               |
| **pH**                             | **6.5 to 8.5**                                                                                         |
| **Conductivity (µmho/cm)**         | **≤ 300 µS/cm** (relaxed up to 750 for some rivers; CPCB allows up to 1500–3000 in polluted stretches) |
| **BOD (mg/L)**                     | **≤ 3.0 mg/L**                                                                                         |
| **Nitrate N (mg/L)**               | **≤ 10 mg/L** (WHO), India prefers **≤ 1.0–2.0** for surface water                                     |
| **Fecal Coliform (MPN/100ml)**     | **≤ 500 MPN/100ml** (drinking), **≤ 2500** (bathing)                                                   |
| **Total Coliform (MPN/100ml)**     | **≤ 500 MPN/100ml**                                                                                    |
| **Fecal Streptococci (MPN/100ml)** | **≤ 100 MPN/100ml** (no firm WHO limit, used in many studies)                                          |


```python
def is_contaminated(row):
    is_cont = 0

    if row["Dissolved Oxygen (mg/L) Min"] < 5.0:
        is_cont+=1
    if row["pH Min"] < 6.5 or row["pH Max"] > 8.5:
        is_cont+=1
    if row["Conductivity (µmho/cm) Max"] > 1500:
        is_cont+=1
    if row["BOD (mg/L) Max"] > 3.0:
        is_cont+=1
    if row["Nitrate N (mg/L) Max"] > 2.0:
        is_cont+=1
    if row["Fecal Coliform (MPN/100ml) Max"] > 2500:
        is_cont+=1
    if row["Total Coliform (MPN/100ml) Max"] > 500:
        is_cont+=1
    if row["Fecal Streptococci (MPN/100ml) Max"] > 100:
        is_cont+=1

    if is_cont >= 3:
      return 1
    else:
      return 0

df["Contaminated"] = df.apply(is_contaminated, axis=1)
```
